Wordclouds. Run in interactive console, do gen_wcloud(typ) where typ is 'INTJ', 'ENFP', etc. Results are in \img folder

```
import pandas as pd
import numpy as np
import feather
import matplotlib.pyplot as plt 
from wordcloud import *

types = ('ENFJ','ENFP','ENTJ','ENTP','ESFJ','ESTJ','ESFP','ESTP','INFP','INFJ','INTJ','INTP','ISFJ','ISFP','ISTJ','ISTP')

df = pd.read_csv('mbti_1_1.csv')
df.posts = df['posts'].str.lower()
df.posts = df['posts'].replace(to_replace=r'[^\w\s]', value='', regex=True).str.split()
wordcloud_strings = {} 
for typ in types:
    print('generating type ' + typ)
    df_typ = df.loc[df['type']==typ]
    str = ''
    for index, row in df_typ.iterrows():
        str += ' '.join(row['posts'])+' '
    wordcloud_strings[typ] = str 
print('generated strings.')

def gen_wcloud(typ):
    wordcloud = WordCloud(width = 800, height = 800).generate(wordcloud_strings[typ])
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.show()
```
