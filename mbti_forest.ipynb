Code below tries to predict the MBTI types, given a one-hot-encoding row for a person's tweets.<br>
The label predictions are strings: \['INTJ', 'ENTP', etc.\] <br>
The accuracy seems pretty low.
____
With abbreviated word list (dictionary-checked):
- n_estimators=100 -> 47.16% accuracy
- n_estimators=500 -> 47.45% accuracy
- n_estimators=1000 -> 47.45% accuracy

With full word list (dictionary-checked):
- n_estimators=100 -> 47.91% accuracy
- n_estimators=500 -> 48.16% accuracy
- n_estimators=1000 -> 48.29% accuracy

With all 200,000 words (not dictionary-checked): <br>
- n_estimators =100 -> 47.4%, =500 -> 48.0%, =1000 -> 47.7%

```
import pandas as pd
import numpy as np
import feather
from sklearn.feature_extraction.text import CountVectorizer

df = feather.read_dataframe('ohe_data_checked.feather') #or whichever onehotencoding table to use
print('loaded dataframe.')
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

from sklearn import metrics 
y = df['type']
x = df.drop(['type'], axis=1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.70, random_state = 5)
print('split dataset.')

def run_forest(num_est=100):
    clf = RandomForestClassifier(n_estimators=num_est)
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x)
    print("accuracy:", metrics.accuracy_score(y, y_pred))

run_forest(100)
```
